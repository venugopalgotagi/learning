# Linear Regression using Boston Housing Data

```python
# Import statements
import logging
import numpy as np
import pandas as pd
import warnings

import seaborn as sns
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D

from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures, StandardScaler

warnings.filterwarnings("ignore")
```

```python
boston_df = pd.read_csv("/kaggle/input/bostonhousingdataset/boston-housing-dataset.csv")
boston_df.info()
```

```python
# Delete None values  and index column

boston_df.isna().sum()

#remove None and  values and first column (index column)
boston_df = boston_df.dropna()

boston_df = boston_df.drop(columns=['Unnamed: 0'],errors='ignore')

boston_df.dtypes
boston_df.describe()
```

```python
# Get correlation coeeficients
crr_boston = boston_df.corr().round(2)
# Print heat map
plt.figure(figsize=(13,13))
sns.heatmap(data = crr_boston, vmin=-1,vmax=1,center=0, annot=True, cmap='vlag')
plt.show()
```

```python
#Show coefficients that are greater than 0.5
abs(crr_boston.MEDV) >= 0.47
logging.info(boston_df.corr().abs().nlargest(4, 'MEDV').index)
```

```python
#print the correlation of columns
plt.scatter(x=boston_df['LSTAT'],y=boston_df['RM'], c='RED')
plt.xlabel('LSTAT')
plt.ylabel('RM')
plt.show()
#print the correlation of columns
plt.scatter(x=boston_df['RM'],y=boston_df['MEDV'],marker='X',c='GREEN')
plt.xlabel('RM')
plt.ylabel('MEDV')
plt.show()

plt.scatter(x=boston_df['LSTAT'],y=boston_df['MEDV'],marker='o',c='ORANGE')
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
plt.show()

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(boston_df['LSTAT'],boston_df['RM'],boston_df['MEDV'],c='GREEN')
plt.show()
```

```python
X = boston_df.copy()
logging.info(X.head())
logging.info("="*100)
y = pd.DataFrame(np.c_[boston_df['MEDV']],columns=['MEDV'])
logging.info(y.head())
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=5,shuffle=True)
X_cv,tmp_test,y_cv,tmp_y = train_test_split(X_test,y_test,test_size=0.2,random_state=5,shuffle=True)
logging.info(f"{X_train.shape}, {y_train.shape}")
logging.info(f"{X_test.shape},{y_test.shape}")
logging.info(f"{X_cv.shape},{y_cv.shape}")
```

```python
scalar = StandardScaler()
pca = PCA(n_components=13)
poly = PolynomialFeatures(degree=2,include_bias=False)

X_train = scalar.fit_transform(X_train)
X_train = pca.fit_transform(X_train)
X_train= poly.fit_transform(X_train)

X_cv = scalar.transform(X_cv)
X_cv = pca.transform(X_cv)
X_cv = poly.transform(X_cv)
```

```python
linear_regression = LinearRegression()
linear_regression.fit(X_train,y_train)

logging.info(linear_regression.coef_)
logging.info(linear_regression.intercept_)
```

```python
y_train_pred = linear_regression.predict(X_train)
logging.info(f"{linear_regression.score(X_train,y_train):.2f}")
logging.info(f"{mean_squared_error(y_train,y_train_pred):.2f}")
plt.scatter(x=y_train, y=y_train_pred,c='GREEN',marker='X')
plt.show()
```

```python
y_cv_pred = linear_regression.predict(X_cv)
logging.info(f"{linear_regression.score(X_cv,y_cv):.2f}")
logging.info(f"{mean_squared_error(y_cv,y_cv_pred):.2f}")
plt.scatter(x=y_train, y=y_train_pred,c='GREEN',marker='X')
plt.show()
```

```python
X_test = scalar.transform(X_test)
X_test = pca.transform(X_test)
X_test = poly.transform(X_test)

y_test_pred = linear_regression.predict(X_test)
logging.info(f"{linear_regression.score(X_test,y_test):.2f}")
logging.info(f"{mean_squared_error(y_test,y_test_pred):.2f}")
plt.scatter(x=y_test, y=y_test_pred,c='GREEN',marker='X')
plt.show()
```
